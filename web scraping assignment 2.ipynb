Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You

jobs data. 
This task will be done in following steps:

2. Enter “Data Analyst” in “Job title, Skills” field and enter “Bangalore” in “enter the  location” field. 
3. Then click the search button. 
4. Then scrape the data for the first 10 jobs results you get. 
5. Finally create a dataframe of the scraped data. 
Note: All of the above steps have to be done in code. No step is to be done manually. 

```python
import selenium
import pandas as pd
from selenium import webdriver 
import warnings
warnings.filterwarnings('ignore')
import time
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.shine.com/")#First get the webpage https://www.shine.com/ 
```

```python
job_title.send_keys('Data Analyst')
```

```python
location = driver.find_element(By.XPATH,'//input[@name="id_loc"]')#enter “Bangalore” in “enter the  location” field
```

```python
location.send_keys('Bangalore')
```

```python
search = driver.find_element(By.XPATH,'//div[@class="searchForm_btnWrap_advance__VYBHN"]/button[1]')#Then click the search button. 
search.click()
```

```python
#Then scrape the data for the first 10 jobs results you get. 
```

```python
job_title = []
title=driver.find_elements(By.XPATH,'//h2[@itemprop="name"]/a')

for i in title[0:10]:
    job_title.append(i.text)
job_title
```

```python
company_name = []
name=driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]/span')

for i in name[0:10]:
    company_name.append(i.text)
company_name
```

```python
experience_required = []
experience_tags = driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
for i in experience_tags[0:10]:
    experience_required.append(i.text)
experience_required
```

```python
df = pd.DataFrame({'JOB_TITLE':job_title,'COMPANY_NAME':company_name,'JOB_LOCATION':job_location,'EXPERIENCE_REQUIRED':experience_required})
df# Finally create a dataframe of the scraped data. 
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART")
```

Q8: Write a python program to scrape data for Top 1000 Quotes of All Time. 
The above task will be done in following steps: 
1. First get the webpage https://www.azquotes.com/ 
2. Click on Top Quotes
3. Than scrap 
a) Quote b) Author c) Type Of Quotes 

```python
import selenium
import pandas as pd
from selenium import webdriver 
import warnings
warnings.filterwarnings('ignore')
import time
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.azquotes.com/")#First get the webpage https://www.azquotes.com/
```

```python
ths_quo = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')#Click on Top Quotes
ths_quo.click()
```

```python
quote = []#Than scrap a) Quote b) Author c) Type Of Quotes
author = []
t_o_q = []
for pages in range(0,10):
    
    qt = driver.find_elements(By.XPATH,'//div[@class="wrap-block"]/p/a[2]')
    for i in qt:
        quote.append(i.text)
        
    at = driver.find_elements(By.XPATH,'//div[@class="wrap-block"]/div/a')
    for i in at:
        author.append(i.text)
        
    to = driver.find_elements(By.XPATH,'//div[@class="tags"]/a')
    for i in to:
        t_o_q.append(i.text)
        
    next_page = driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[12]/a')
    next_page.click()
```

```python

grouped_data = [t_o_q[i:i+3] for i in range(0, len(t_o_q), 3)]
for group in grouped_data:
    print(group)
```

```python
df = pd.DataFrame({'QUOTE':quote,'AUTHOR':author,'TYPE_OF_QUOTES':grouped_data})
df
```
