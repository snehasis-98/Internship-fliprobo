# 1. Scrape  the  details  of  most  viewed  videos  on  YouTube  from  Wikipedia.  
## Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: 
A) Rank   
B) Name   
C) Artist   
D) Upload date   
E) Views   

```python
import selenium
import pandas as pd

from selenium import webdriver 
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
from selenium.webdriver.support import expected_conditions as EC

import warnings
warnings.filterwarnings('ignore')
import time
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python

tables=[]


        
try:
    table = driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr') 
    for i in table:
        row = [item.text for item in i.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td')]
        tables.append(row)
except:
    tables.append("-")
        
tables 
```

```python
len(tables)
```

```python
split_list = [','.join(item).split(",") for item in tables]
print(split_list)
```

```python
lists = split_list[1]
```

```python
grouped_daa = [lists[i:i+6] for i in range(0, len(lists), 6)]
for group in grouped_daa:
    print(group)
```

```python
columns = ['Name' ,'Artist' ,'Views(Billions)' ,'Upload date' ,'Upload year','Notes']
df = pd.DataFrame(grouped_daa, columns=columns)
df
```

```python
df.index = pd.RangeIndex(start=df.index.min() + 1, stop=df.index.max() + 2)
```

```python
df = df.rename_axis('Rank')
df
```
